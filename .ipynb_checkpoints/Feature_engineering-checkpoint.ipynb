```python
def remove_puns(text):
    return text.translate(str.maketrans('','',punctuations))
```

```python
import contractions
import re

# Dictionary for informal/missing apostrophes
informal_fixes = {
    "im": "i am",
    "dont": "do not",
    "cant": "can not",
    "wont": "will not",
    "didnt": "did not",
    "isnt": "is not",
    "wasnt": "was not",
    "wouldnt": "would not",
    "couldnt": "could not",
    "shouldnt": "should not",
    "hasnt": "has not",
    "havent": "have not",
    "hadnt": "had not",
    "thats": "that is",
    "whats": "what is",
    "heres": "here is",
    "theres": "there is"
}

def clean_review(text):
    # First, fix informal contractions (without apostrophes)
    for old, new in informal_fixes.items():
        pattern = r'\b' + old + r'\b'
        text = re.sub(pattern, new, text, flags=re.IGNORECASE)
    # Then, fix standard contractions (with apostrophes)
    text = contractions.fix(text)
    return text
```

```python
imdb_data['review'] = imdb_data['review'].apply(clean_review)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopword(text):
    new_text = []
    for i in text.split():
        if i not in stopwords.words('english'):
            new_text.append(i)
    return " ".join(new_text)
```

```python
imdb_data['review'] = imdb_data['review'].apply(remove_stopword)
```
